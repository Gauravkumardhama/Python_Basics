```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(10)
y=x**3
print(x)
print(y)
plt.plot(x,y)
```

```python
plt.scatter(x,y,marker='^',color='red',label="Points")
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Intersection')
plt.show()
```

```python
plt.plot(x,x,label="x")
plt.plot(x,x**3,label="x^3")
plt.legend()    #for label
plt.show()
```

```python
#bar charts
no=np.random.randint(10,20,5)
indices=np.arange(5)

no2=np.random.randint(10,20,5)
indices2=np.arange(5)
plt.bar(indices,no,0.25,color='yellow',label='gold')
plt.bar(indices2+0.25,no2,0.25,color='silver',label='silver')
plt.legend()
plt.show()
```

```python
#pie chart
labels=["a","b","c","d"]
value=[10,30,5,40]
plt.pie(value,labels=labels,radius=2,explode=(0,0,1,0),shadow=True)
plt.show()
```
